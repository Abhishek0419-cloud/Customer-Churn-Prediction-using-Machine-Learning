```python
import pickle
import pandas as pd
```

```python
# load saved dictionary
saved_data = pickle.load(open("customer_churn_model.pkl", "rb"))

# extract model and training feature names
model = saved_data["model"]
feature_names = saved_data["features_names"]

print("Model loaded successfully!")
print("Total features model expects:", len(feature_names))
```

```python
# NEW CUSTOMER DATA (RAW INPUT)
data = {
    'tenure': [12],
    'MonthlyCharges': [85.5],
    'TotalCharges': [1020.5],
    'Contract': [0],
    'PaperlessBilling': [1]
}

customer_df = pd.DataFrame(data)
customer_df
```

```python
# create empty dataframe with same columns as training data
input_df = pd.DataFrame(columns=feature_names)

# create one empty row
input_df.loc[0] = 0

# copy available customer values into correct columns
for col in customer_df.columns:
    if col in input_df.columns:
        input_df[col] = customer_df[col]

input_df.head()
```

```python
prediction = model.predict(input_df)
probability = model.predict_proba(input_df)[0][1]

if prediction[0] == 1:
    print("Customer is likely to CHURN")
else:
    print("Customer is likely to STAY")

print("Churn Probability:", round(probability*100,2), "%")
```
